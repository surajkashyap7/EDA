```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from datetime import datetime
import requests
```

```python
data = pd.read_csv("D:\\Projects\\covid_19_india.csv")
```

```python
data.head()
```

```python
# change some columns name
data = data.rename(columns={'State/UnionTerritory':'States', 'Cured':'Recovery'})
```

```python
data.head()
```

```python
# drop some columns
data = data.drop(['Sno', 'Time', 'ConfirmedIndianNational', 'ConfirmedForeignNational'],  axis=1) # axis =1 (column)
```

```python
data.head()
```

```python
# find active case
data['Active'] = data['Confirmed'] - data['Recovery'] - data['Deaths']  # 1-0-0 = 1
```

```python
data.head()
```

```python
data = data.sort_values(['Date', 'States']).reset_index(drop=True)
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data
```

```python
india_cases = data[data['Date'] == data['Date'].max()].copy().fillna(0)
india_cases.index = india_cases['States'] # index = states
india_cases = india_cases.drop(['States', 'Date'], axis=1)

india_cases
```

```python
df = pd.DataFrame(pd.to_numeric(india_cases.sum())).transpose()
df.style.background_gradient(cmap='summer_r', axis=1)
```

```python
india_cases.sort_values('Active', ascending = False).style\
    .background_gradient(cmap='BrBG_r', subset=['Confirmed'])\
    .background_gradient(cmap='BuGn_r', subset=['Deaths'])\
    .background_gradient(cmap='BuPu_r', subset=['Recovery'])\
    .background_gradient(cmap='Oranges_r', subset=['Active'])
```

```python
def horizontal_bar_chart(df, x, y, title, x_label, y_label, color):
    fig = px.bar(df, x=x, y=y, orientation='h', title=title, labels={x.name=x_label, y.name=y_label, color_discreate_sequence=[color]})
    fig.update_layout(yaxis={'Categoryorder':'total ascending'})
    fig.show()
```
