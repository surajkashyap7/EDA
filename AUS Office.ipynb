```python
Fan_Type=df["Fan type"].value_counts().sort_values().reset_index()
Fan_Type
```

```python
ReFanSpeed=df[["Month","Re_Fan_Speed_CMD"]].value_counts().reset_index()
ReFanSpeed.columns=["Month","Re_Fan_Speed_CMD","number_of_speed"]
ReFanSpeed
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x="Month",y="Re_Fan_Speed_CMD", data=ReFanSpeed)
plt.title("Re Fan Speed with Month",fontsize=20)
plt.show()
```
